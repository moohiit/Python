1. Now from the churn data frame, try to sort the data by the tenure column according to the 
descending order.

```python
import pandas as pd

df=pd.read_csv('customer_churn.csv')
df=df.sort_values(by=['tenure'],ascending=False).reset_index(drop=True)
df
```

  
2. Fetch all the records that are satisfying the following condition:  
a. Tenure>50 and the gender as ‘Female’.
b. Gender as ‘Male’ and SeniorCitizen as 0.  
c. TechSupport as ‘Yes’ and Churn as ‘No’.  
d. Contract type as ‘Month-to-month’ and Churn as ‘Yes’. 

```python
# A. Tenure>50 and the gender as 'female'.
filter1=(df['tenure']>50)&(df['gender']=='Female')
df[filter1]
```

```python
#b. Gender as ‘Male’ and SeniorCitizen as 0
filter2=(df['gender']=='Male')&(df['SeniorCitizen']==0)
df[filter2]
```

```python
#c. TechSupport as ‘Yes’ and Churn as ‘No’.
filter3=(df['TechSupport']=='Yes')&(df['Churn']=='No')
df[filter3]
```

```python
#d. Contract type as ‘Month-to-month’ and Churn as ‘Yes’.
filter4=(df['Contract']=='Month-to-month')&(df['Churn']=='Yes')
df[filter4]
```
