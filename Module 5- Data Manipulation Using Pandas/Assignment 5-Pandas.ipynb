1. Use a for loop to calculate the number of customers that are getting the tech support and 
are male senior citizen. 

```python
import pandas as pd
df=pd.read_csv('customer_churn.csv')
df.head()
```

```python
#the number of customers that are getting the tech support and are male senior citizen. 
filter1=(df['SeniorCitizen']==1)&(df['gender']=='Male')&(df['TechSupport']=='Yes')
```

```python
#Using for Loop find the no of customers that follows the above filter:
sum=0
for i in range(len(df[filter1])):
    sum+=1
print('Sum of costumer:',sum)
```
