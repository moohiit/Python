- Using pandas import the dataset as dataframe 

```python
data=pd.read_csv("E:\\Data Science Architect Master's Course\\Unit 5-Python Basics\\Module-9 Unsupervised Learning\\store_data.csv")
data.head()
```

```python
data.fillna('Nothing',inplace=True)
data.head()
```

-  Install the mixtend library to use apriory and association rule mining 

```python
from apyori import apriori
```

- Using the apriori algorithm generate a list of item frequently brought together. 

```python
transaction=[]
for i in range(1,data.shape[0]):
    transaction.append([str(data.values[i,j]) for j in range(0,data.shape[1])])
```

- Generate the association rules for the given items from apriori algorithm 

```python
association_rules = apriori(transaction, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
for i in range(0, len(association_results)):
    print(association_results[i][0])
```
