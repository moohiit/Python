```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import load_iris
```

- From the dataset extract the data property. 

```python
iris=load_iris()
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],columns= iris['feature_names'] + ['target'])
print(iris.feature_names,iris.target_names)
data.head()
```

```python
iris=load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data['target']=pd.Series(iris.target)
data.head()
```

```python
#Visualising the data:
sns.pairplot(data)
```

- Train an AgglomerativeClustring model based on the data. 

```python
X=data.iloc[:,[0,2]].values
```

```python
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
labels
```

- Plot dendrogram to visualize the clustering linkage 

```python
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

```python
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=20, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=30, marker='o', color='blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=40, marker='o', color='green')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, marker='o', color='purple')
plt.scatter(X[labels==4, 0], X[labels==4, 1], s=10, marker='o', color='orange')
plt.show()
```

# Example of dendrogram:

```python
from scipy.cluster import hierarchy 
import matplotlib.pyplot as plt
import numpy as np
#A very basic example:

ytdist = np.array([662., 877., 255., 412., 996., 295., 468., 268.,400., 754., 564., 138., 219., 869., 669.]) 
Z = hierarchy.linkage(ytdist, 'single')
plt.figure()
dn = hierarchy.dendrogram(Z)
#Now, plot in given axes, improve the color scheme and use both vertical and horizontal orientations:

hierarchy.set_link_color_palette(['m', 'c', 'y', 'k'])
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
dn1 = hierarchy.dendrogram(Z, ax=axes[0], above_threshold_color='y',
                            orientation='top')
dn2 = hierarchy.dendrogram(Z, ax=axes[1],
                            above_threshold_color='#bcbddc',
                            orientation='right')
hierarchy.set_link_color_palette(None)  # reset to default after use
plt.show()
```

```python
#Examples
>>> from scipy.cluster.hierarchy import dendrogram, linkage
>>> from matplotlib import pyplot as plt
>>> X = [[i] for i in [2, 8, 0, 4, 1, 9, 9, 0]]
>>> Z = linkage(X, 'ward')
>>> fig = plt.figure(figsize=(8, 5))
>>> dn = dendrogram(Z)
>>> Z = linkage(X, 'single')
>>> fig = plt.figure(figsize=(10, 8))
>>> dn = dendrogram(Z)
>>> plt.show()
```
