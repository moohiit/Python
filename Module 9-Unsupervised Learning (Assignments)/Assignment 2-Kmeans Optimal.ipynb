```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
```

- Load customer data. 

```python
data=pd.read_csv("E:\\Data Science Architect Master's Course\\Unit 5-Python Basics\\Module-9 Unsupervised Learning\\customers.csv")
data.head()
```

- Check the number of cells in each column with null values. 

```python
data.isnull().sum()
```

```python
data.info()
```

- Create a scatter plot with Age as X and Spending Score as Y. 

```python
X=data['Age']
Y=data['Spending Score (1-100)']

# Creating a sctter plot:
sns.scatterplot(X,Y)
```

```python
# Creating a Scatter plot to data:
sns.pairplot(data)
```

- Find out the best number for clusters between 1 and 10 (inclusive) using the elbow method. 

```python
N=range(1,10)
df=np.array(list(zip(X,Y)))

kmean=[KMeans(n_clusters=i) for i in N]
score=[kmean[i].fit(df).score(df) for i in range(len(kmean))]

plt.figure(figsize=(10,10))
plt.plot(N,score)
plt.xlabel("Number of Clusters")
plt.ylabel("Score")
#plt.yticks(N)
plt.title("Elbow Curve")
plt.show()
```

```python
# From the above graph we conclude that optimal value of cluster is '4':
```

- Draw a scatter plot displaying data points colored on the basis of clusters. 

```python
from sklearn.cluster import KMeans
import numpy as np

x=data['Age']
y=data['Spending Score (1-100)']

n=4
df=np.array(list(zip(x,y)))

# Creating clusters:
kmeans=KMeans(n_clusters=n)

# Fitting the data on Model
kmeans.fit(df)
# Centroids of clusters:
centriods=kmeans.cluster_centers_
# Labels:
labels=kmeans.labels_
#Difing the colors:
color=['r','g','b','y','m']

# Visualising the data:
import matplotlib.pyplot as plt
fig, ax= plt.subplots()
for i in range(n):
    points=np.array([df[j] for j in range(len(df)) if labels[j]==i])
    ax.scatter(points[:,0],points[:,1],s=8,c=color[i])
ax.scatter(centriods[:,0],centriods[:,1],marker='*',c='black')
```
