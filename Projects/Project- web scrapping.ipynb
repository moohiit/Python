```python
#  Importing the library:
import requests

#specifying the URL:
url_link="https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
html=requests.get(url_link).text
```

```python
html
```

```python
# Importing Beautiful soup for extracting html:
from bs4 import BeautifulSoup

soup=BeautifulSoup(html,'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
soup.find_all("a")
```

```python
all_link=soup.find_all("a")

for link in all_link:
    print(link.get("href"))
```

```python
all_table=soup.find_all("table")
print(all_table)
```

```python
right_table=soup.find("table",class_="wikitable sortable")
right_table
```

```python
table_link=right_table.find_all('a')
table_link
```

```python
country=[]
for link in table_link:
    country.append(link.get("title"))

country
```

```python
import pandas as pd

df=pd.DataFrame(country,columns=['Country'])
```
