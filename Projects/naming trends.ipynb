Importing Dataset from the all names files in list:

```python
years = []
for year in range(1880,2021):
    years.append(pd.read_csv(f'yob{year}.txt',names = ['Name','Sex','Babies']))
    years[-1]['Year'] = year
```

```python
years
```

Converting all the data into a single dataframe using concatenate function:

```python
df=pd.concat(years)
df.head()
```

# Create a pivot table of Babies as male and female columns with index as year

```python
name_df=pd.pivot_table(data=df, index=['Year'], columns=['Sex'], values=['Babies'],aggfunc='sum')
name_df
```

```python

name_df.plot(kind='barh', figsize=(18,60));
plt.show()
```

Sorting the data according to number of babies:

```python
pd.options.display.max_rows=100
sort_babies_names=df.sort_values(by='Babies',ascending=False).reset_index(drop=True)
sort_babies_names.head(100)
```

```python
pd.options.display.max_rows=100
top_100_names=sort_babies_names[['Name','Babies','Sex']].groupby('Name').sum().sort_values(by="Babies",ascending=False)
top_100_names=top_100_names.head(100)
top_100_names.head(100)
```

```python
top_100_names.plot(kind='barh',figsize=(10,60))
plt.show()
```
