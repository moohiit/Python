1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame. 


```python
# Importing libraries:
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Importing Data:
cars=pd.read_csv('mtcars.csv')
cars.rename(columns={'Unnamed: 0':'model'},inplace=True)
cars.head()
```

```python
#Creating Data:
model=cars['model']
cyl=cars['cyl']
```

```python
#Ploting Histogram:
plt.figure(figsize=(20,20))
plt.pie(cyl,labels=model,autopct='%0.2f%%',startangle=30,shadow=True)
plt.title('Pie Chart Of Cylinder')
plt.show()
```
