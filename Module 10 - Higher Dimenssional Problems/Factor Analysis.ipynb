## Kaiser-Meyer-Olkin (KMO) Test

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df)
kmo_model
```

## Common Factor Analysis

```python
fa = FactorAnalyzer(n_factors=df.shape[1]-1)
fa.fit(df)
ev,v = fa.get_eigenvalues()
```

```python
min(ev)
```

```python
loading_vector = pd.DataFrame(fa.loadings_,index=df.columns)
```

```python
loading_vector.head()
```

## Scree Plot

```python
import numpy as np

ax = plt.subplot()
ax.scatter(range(1,df.shape[1]+1),ev)
ax.plot(range(1,df.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
x_ticks = [i for i in range(1, df.shape[1]-1)]
y_ticks = [i for i in np.arange(ev.min(), ev.max())]
ax.set_xticks(x_ticks)
plt.grid()
plt.show()
```
