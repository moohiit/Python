```python
import pandas as pd
```

```python
# Pandas Merge
first_df = pd.DataFrame({
    'name': ['Jane', 'John', 'Jacob'],
    'age': [15, 25, 25]
}, index=[0, 1, 2])

second_df = pd.DataFrame({
    'name': ['June', 'Julia', 'Julie'],
    'age': [28, 18, 20]
}, index=[3, 4, 5])

third_df = pd.DataFrame({
    'name': ['Jean', 'Jenny', 'Jessie'],
    'age': [17, 28, 30]
}, index=[6, 7, 8])

pd.concat([first_df, second_df, third_df])
```

```python
# Pandas Join

df1 = pd.DataFrame({'key': ['0', '1', '2', '3'],
                   'A': ['A0', 'A1', 'A2', 'A3']})
df2 = pd.DataFrame({'key': ['0', '1'],
                   'A': ['A0', 'A1',]})
df1.join(df2, lsuffix='_first_df', rsuffix='_second_df')
```

```python
# Pandas Concat

df1 = pd.DataFrame({'left_key': ['1', '2', '3', '4'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'right_key': ['1', '2', '3', '4'],
                    'value': [5, 6, 7, 8]})
df1.merge(df2, left_on='left_key', right_on='right_key')
```

```python
df1.join(df2, lsuffix='_df1',rsuffix='_df2')
```
